```python
# import the NABR_history.csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

historic= pd.read_csv('./NABR_historic.csv')
nearterm= pd.read_csv('./nearterm_data_2020-2024.csv')
combined= pd.concat([historic, nearterm], axis=0)
# # check the data shape
# print(historic.shape)
# print(nearterm.shape)

# #check null values
# print(historic.isnull().sum())
# print(nearterm.isnull().sum())
print(combined.isnull().sum())
print(combined.shape)
# # #check the data types
# print(historic.dtypes)

# print(combined.TimePeriod.unique())

# Exclude non-numeric columns for correlation calculation
numeric_columns = combined.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr()

# Plot the correlation heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
